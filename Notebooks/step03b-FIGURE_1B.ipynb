```python
k = stats.gaussian_kde( coeff_list.coeff)
```

```python
xx = linspace(-4,5,1000)
```

```python
figure( figsize=(4,2))
plot( xx, k.evaluate(xx) )
xlabel('Coefficient')
ylabel('Density')
savefig('./Figures/Fig_1_B.pdf')
savefig('./Figures/Fig_1_B.png')
show()
```
